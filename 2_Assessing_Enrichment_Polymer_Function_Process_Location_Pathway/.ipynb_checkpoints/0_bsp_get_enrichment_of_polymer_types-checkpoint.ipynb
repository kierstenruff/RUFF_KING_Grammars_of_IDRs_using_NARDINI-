# Part 2: Get enrichment by polymer type

## 2.1 - Define polymer types

```python
ncprzscore=3

polyelectrolytespos=[]
polyelectrolytesneg=[]
polyampholytesmixed=[]
polyampholytesblocky=[]
associativepolymersaro=[]
associativepolymershyd=[]
polartracts=[]
polartractsH=[]
polartractsN=[]
polartractsQ=[]
polartractsS=[]
polartractsT=[]

for s in range(0,len(subseqs)):
    # Get current z-score vector
    currzvec=subzvecaa[s,:]
    
    
    # Define postive polyelectrolytes
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])>0.35 and (currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])>0.35:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polyelectrolyte pos')
        #print(subseqs[s])
        #print((currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')]))
        #print((currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')]))
        polyelectrolytespos.append(subaccswnums[s])
        
    # Define negative polyelectrolytes
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])>0.35 and (currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<-0.35:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polyelectrolyte neg')
        #print(subseqs[s])
        #print((currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')]))
        #print((currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')]))
        polyelectrolytesneg.append(subaccswnums[s])

    # Define polyampholytes - mixed
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])>0.35 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<=0.35 and currzvec[myfeat.index('pos-neg')]<-1:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polyampholyte well-mixed')
        #print(subseqs[s])
        #print((currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')]))
        #print((currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])) 
        polyampholytesmixed.append(subaccswnums[s])

    # Define polyampholytes - blocky
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])>0.35 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<=0.35 and currzvec[myfeat.index('pos-neg')]>1:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polyampholyte blocky')
        #print(subseqs[s])
        #print((currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')]))
        #print((currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])) 
        polyampholytesblocky.append(subaccswnums[s])     
        
    # Define associative polymer aro
    if currzvec[myfeat.index('aro-aro')]<-1 and (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])<0.25 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<0.25:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', associative polymer aro')
        #print(subseqs[s])
        #print((currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')]))
        #print((currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')]))
        associativepolymersaro.append(subaccswnums[s])
        
    # Define associative polymer hyd
    if currzvec[myfeat.index('hyd-hyd')]<-1 and (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])<0.25 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<0.25:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', associative polymer hyd')
        #print(subseqs[s])
        #print((currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')]))
        #print((currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')]))
        associativepolymershyd.append(subaccswnums[s])
        
    # Define polar tracts
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])<0.25 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<0.25 and currzvec[myfeat.index('pol-pol')]>1 and (currzvec[myfeat.index('H Patch')]>1 or currzvec[myfeat.index('Q Patch')]>1 or currzvec[myfeat.index('N Patch')]>1 or currzvec[myfeat.index('T Patch')]>1 or currzvec[myfeat.index('S Patch')]>1):
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polar tract')
        #print(subseqs[s])
        #print((currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')]))
        #print((currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')]))
        polartracts.append(subaccswnums[s])
        
    # Define polar tracts H
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])<0.25 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<0.25 and currzvec[myfeat.index('pol-pol')]>1 and currzvec[myfeat.index('H Patch')]>1:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polar tract H')
        #print(subseqs[s])
        polartractsH.append(subaccswnums[s])
        
    # Define polar tracts N
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])<0.25 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<0.25 and currzvec[myfeat.index('pol-pol')]>1 and currzvec[myfeat.index('N Patch')]>1:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polar tract N')
        #print(subseqs[s])
        polartractsN.append(subaccswnums[s])
        
    # Define polar tracts Q
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])<0.25 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<0.25 and currzvec[myfeat.index('pol-pol')]>1 and currzvec[myfeat.index('Q Patch')]>1:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polar tract Q')
        #print(subseqs[s])
        polartractsQ.append(subaccswnums[s])
        
    # Define polar tracts S
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])<0.25 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<0.25 and currzvec[myfeat.index('pol-pol')]>1 and currzvec[myfeat.index('S Patch')]>1:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polar tract S')
        #print(subseqs[s])
        polartractsS.append(subaccswnums[s])
        
    # Define polar tracts T
    if (currzvec[myfeat.index('FCR')]*stdvals[myfeatcomp.index('FCR')]+meanvals[myfeatcomp.index('FCR')])<0.25 and abs(currzvec[myfeat.index('NCPR')]*stdvals[myfeatcomp.index('NCPR')]+meanvals[myfeatcomp.index('NCPR')])<0.25 and currzvec[myfeat.index('pol-pol')]>1 and currzvec[myfeat.index('T Patch')]>1:
        #print('>'+subgeneswnums[s]+': Cluster '+str(subclusters[s])+', polar tract T')
        #print(subseqs[s])
        polartractsT.append(subaccswnums[s])
```

## 2.2 - Extract enrichment values

```python
accsbypolymertype=[polyelectrolytespos, polyelectrolytesneg, polyampholytesmixed, polyampholytesblocky, associativepolymersaro, associativepolymershyd, polartracts, polartractsH, polartractsN, polartractsQ, polartractsS, polartractsT]
namesbypolymertype=['Cationic polyelectrolytes','Anionic polyelectrolytes','Well-mixed polyampholytes','Blocky polyampholytes','Aromatic associative polymers','Hydrophobic associative polymers','Polar tracts','H polar tracts','N polar tracts','Q polar tracts','S polar tracts','T polar tracts']

numkclusts=max(subclusters)+1

pvalsbylocation=np.zeros((len(accsbypolymertype),numkclusts))
pvalsbylocationlog=np.zeros((len(accsbypolymertype),numkclusts))

numIDRsbypolymertype=[]
for a in range(0,len(accsbypolymertype)):
    print(namesbypolymertype[a]+': '+str(len(accsbypolymertype[a])))
    numIDRsbypolymertype.append(len(accsbypolymertype[a]))
    sumallidrs=0
    for c in range(0,numkclusts):
        print('Cluster_'+str(c))
        midx = [i for i, x in enumerate(subclusters) if x == c]
        
        topaccs=[]
        for i in midx:
            topaccs.append(subaccswnums[i])
        
        print('Number of IDRs: '+str(len(topaccs)))
    
        # For each location determine if that feature is enriched there
        subfeatureyes=0
        subfeatureno=0
        restfeatureyes=0
        restfeatureno=0
        for s in range(0,len(subaccs)):
            if subaccswnums[s] in accsbypolymertype[a]:
                if subaccswnums[s] in topaccs:
                    subfeatureyes=subfeatureyes+1
                    sumallidrs=sumallidrs+1
                else:
                    subfeatureno=subfeatureno+1
            else:
                if subaccswnums[s] in topaccs:
                    restfeatureyes=restfeatureyes+1
                else:
                    restfeatureno=restfeatureno+1

        print(subfeatureyes+subfeatureno+restfeatureyes+restfeatureno)

        # defining the table
        if subfeatureyes+restfeatureyes>0:
            data = [[subfeatureyes, subfeatureno], [restfeatureyes, restfeatureno]]
            print(data)
            # One-sided greater used to test if greater occurrence in nucleolus
            res = fisher_exact(data, alternative='greater')
            print(res[1])
            pvalsbylocation[a,c]=res[1]
            if res[1]>0:
                pvalsbylocationlog[a,c]=-math.log10(res[1])
            else:
                pvalsbylocationlog[a,c]=500
            
    print('Number of IDRs in '+namesbypolymertype[a]+' '+str(subfeatureyes+subfeatureno))
    print(sumallidrs)

print(pvalsbylocation)    
```
